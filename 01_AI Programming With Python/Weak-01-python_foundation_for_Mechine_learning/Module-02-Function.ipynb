## 2. Different Types of Arguments

Functions can accept values in different ways. Here we will learn about positional arguments, variable-length arguments, default arguments, and keyword arguments.

```python
def show_three_values(a, b, c):
    print(f"a = {a}, b = {b}, c = {c}")

show_three_values(10, 30, 20)
```

```python
def show_all_values(*args):
    for value in args:
        print(value)

show_all_values(10, 30, 20, 40)
```

```python
def show_default_values(a=10, b=30, c=10):
    print(f"a = {a}, b = {b}, c = {c}")

show_default_values()
```

```python
def show_keyword_values(**kwargs):
    for key, value in kwargs.items():
        print(f"{key}: {value}")

show_keyword_values(a=10, b=30, c=40, n_estimator=50)
```

## 3. Return Values

A function can either print a result or return it. Returning a value is more useful when we want to use that result later in our program.

```python
def welcome_message():
    print("Hello, user!")

print(type(welcome_message()))
```

```python
def add_numbers(a, b):
    return a + b

print(type(add_numbers(10, 20)))
```

```python
def add_numbers(a, b):
    return a + b

print(type(add_numbers(10.5, 20)))
```

```python
def welcome_message(user="student"):
    return f"Hello, {user}!"

print(type(welcome_message()))
```

## 4. Iterators and Generators

Iterators let us step through values one at a time. Generators are a memory-friendly way to produce values only when needed.

```python
values = {10, 20, 30, 40, 50}
value_iterator = iter(values)

print(next(value_iterator))
print(next(value_iterator))
print(next(value_iterator))
print(next(value_iterator))
print(next(value_iterator))

print("All values in the set:")
for value in values:
    print(value)
```

## 5. Generators

A generator uses `yield` to create values one at a time. This is helpful when we work with large data because it avoids loading everything into memory at once.

```python
data = [x for x in range(500)]  # Example dataset

def data_loader(chunk_size, data):
    for start in range(0, len(data), chunk_size):
        yield data[start:start + chunk_size]

loader = data_loader(5, data)

print(next(loader))
print(next(loader))
```

## 6. Lambda, Map, and Apply

Lambda functions are short, anonymous functions. They are often used with `map()` or Pandas `apply()` when we need a small transformation.

```python
import pandas as pd

# Convert temperature values from Fahrenheit to Celsius.
df = pd.DataFrame({'temperature (F)': [32, 212, 98.6, 68]})
df

df['temperature (C)'] = df['temperature (F)'].apply(lambda x: (x - 32) * 5 / 9)
df
```

```python
def pair_values(a, b):
    return [a, b]

print(type(pair_values(10.5, 20)))

# Unpack the returned list into two variables.
x, y = pair_values(10.5, 20)
print(y)
```

```python
import pandas as pd

# Example of transforming a column with lambda and apply.
df = pd.DataFrame({'A': [1, 2, 3, 4], 'B': [5, 6, 7, 8]})
df

df['A'] = df['A'].apply(lambda x: x ** 0.5)
df
```

```python
import pandas as pd

# Create a simple feature from a numeric column.
df = pd.DataFrame({'total_spent': [120, 600, 450, 800]})
df['is_high_value'] = df['total_spent'].apply(lambda x: 1 if x > 500 else 0)
df
```
