## 1. Variables and Data Types

In Python, a variable is used to store data. Python is dynamically typed, which means the type of a variable is decided automatically based on the value assigned to it.

In this section, we will see:
- integer values
- float values
- string values
- input handling
- type conversion

```python
# Store an integer in a variable
a = 10

print("Value of a:", a)
print("Type of a:", type(a))
```

```python
# Store a floating-point number in a variable
b = 10.5

print("Value of b:", b)
print("Type of b:", type(b))
```

```python
# A variable can hold a new value of a different type
a = b

print("Value of a:", a)
print("Type of a:", type(a))
```

```python
# A string variable
name = "phitron"

print("Value of name:", name)
print("Type of name:", type(name))
```

```python
# Input() always returns a string
data = input()

print("Input value:", data)
print("Type of data:", type(data))
```

```python
# Convert a string input to an integer
a = input()
a = int(a)

print("Converted value:", a)
print("Type of a:", type(a))
```

```python
# Read two numbers from one line of input
row_number = input()

numbers = row_number.split()

a = int(numbers[0])
print(a)
print(type(a))

b = int(numbers[1])
print(b)
print(type(b))

print(a + b)
```

## 2. Conditional Statements

Conditional statements help the program make decisions. In Python, we use `if`, `elif`, and `else` to run different code depending on a condition.

```python
import torch
import torch.nn as nn

# Example input: a single sample with 4 features.
image_features = torch.tensor([[0.15, 0.85, 0.25, 0.70]])

# A simple feed-forward network for a binary classification demo.
model = nn.Sequential(
    nn.Linear(4, 12),
    nn.ReLU(),
    nn.Linear(12, 1),
    nn.Sigmoid()
)

# Get prediction without training the model.
with torch.no_grad():
    prediction = model(image_features)

prob = prediction.item()
print(f"Predicted probability: {prob:.4f}")

# Interpret the result for learning purposes.
if prob == 0.5:
    print("The model is uncertain.")
elif prob > 0.5:
    print("Predicted class: Positive")
else:
    print("Predicted class: Negative")
```

## 3. Loops and Lists

A loop lets us repeat a task multiple times, while a list helps us store values in order. In this section, we use both to check how the KNN model performs for different values of `k`.

```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier

# Load the dataset and split it into training and test sets.
data = load_breast_cancer()
X, y = data.data, data.target

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

# Test different values of k and store the accuracy for each one.
k_values = range(1, 51)
accuracies = []
results = []

for k in k_values:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)

    score = knn.score(X_test, y_test)
    accuracies.append(score)
    results.append((k, score))

print("K Value | Test Accuracy")
for k, score in results:
    print(f"{k:7d} | {score:.4f}")

mx_accuracy = max(accuracies)
best_k_index = accuracies.index(mx_accuracy)
best_k = k_values[best_k_index]

print(f"Best accuracy: {mx_accuracy:.4f}")
print(f"Best k: {best_k}")
```
