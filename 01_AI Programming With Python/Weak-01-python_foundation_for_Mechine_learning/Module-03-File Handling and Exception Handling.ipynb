# File Handling and Exception Handling in Python

This notebook is a beginner-friendly guide to working with files and handling errors in Python.

## Learning goals
- Read text files in different ways
- Write and append data to files
- Understand file pointer movement with `tell()` and `seek()`
- Solve simple file-practice problems
- Handle runtime errors safely using `try`, `except`, `else`, and `finally`

Each example is short and written so a student can follow it step by step.

```python
from pathlib import Path

data_dir = Path("./sample_data")
data_dir.mkdir(parents=True, exist_ok=True)

sample_path = data_dir / "sample.txt"
test_path = data_dir / "test.txt"
test2_path = data_dir / "test2.txt"
counter_path = data_dir / "counter_of_string.txt"
write_read_path = data_dir / "write_read.txt"

if not sample_path.exists() or sample_path.stat().st_size == 0:
    sample_path.write_text(
        "Python helps us automate tasks.\n"
        "File handling stores data in text files.\n"
        "We can read and write files easily.\n",
        encoding="utf-8",
    )

print(sample_path.resolve())
```

## 1. File Read

Reading a file means opening it and getting its content into Python. In this section, we will read a file using `read()`, `readlines()`, and a `with` block.

```python
with open(sample_path, "r") as file:
    content = file.read()

print(type(content))
print(content)
print(file.closed)
```

```python
with open(sample_path, "r", encoding="utf-8") as file:
    content = file.readlines()

content = list(map(str.strip, content))
clean_content = list(filter(lambda line: len(line) >= 1, content))
print(clean_content)
```

```python
with open(sample_path, "r", encoding="utf-8") as file:
    content = file.readlines()
    print(content)

print(file.closed)
```

```python
with open(sample_path, "r", encoding="utf-8") as file:
    for line in file:
        print(line.strip())
```

## 2. File Write

Writing a file lets us save new text, update existing text, or add more text to the end of a file.

```python
# Create a new text file and write two lines into it.
with open(test_path, "w", encoding="utf-8") as file:
    file.write("Hello test file\n")
    file.write("How is learning going on?")
```

```python
# Opening a file in 'w' mode overwrites the previous content.
with open(test_path, "w", encoding="utf-8") as file:
    file.write("Second attempt\n")
    file.write("Learning is going fast\n")
```

```python
# Append mode adds new text at the end of the file.
with open(test_path, "a", encoding="utf-8") as file:
    file.write("The learning of AI/ML is fun\n")
    file.write("I am enjoying the classes\n")
```

```python
strings = ["hello\n", "hi\n", "good bye\n", "what up\n"]

with open(test2_path, "a", encoding="utf-8") as file:
    file.writelines(strings)
```

```python
with open(sample_path, "r", encoding="utf-8") as file:
    print(file.tell())
    print(file.read())
    print(file.tell())
    print(file.read())
```

```python
with open(sample_path, "r", encoding="utf-8") as file:
    print(file.tell())
    print(file.read(5))
    print(file.tell())
    print(file.read())
    print(file.tell())
```

```python
with open(sample_path, "r", encoding="utf-8") as file:
    print(file.tell())
    print(file.read(5))
    print(file.tell())
    file.seek(0)
    print(file.tell())
    print(file.read(5))
```

## 3. Practice 1

Read `sample.txt`, count the number of lines, words, and characters, and save the result into `counter_of_string.txt`.

```python
with open(sample_path, "r", encoding="utf-8") as file:
    lines = file.readlines()

line_count = len(lines)
word_count = sum(len(line.split()) for line in lines)
char_count = sum(len(line.strip().replace(" ", "")) for line in lines)

summary = (
    f"total line: {line_count}\n"
    f"total number of words: {word_count}\n"
    f"total number of char: {char_count}"
)

with open(counter_path, "w", encoding="utf-8") as file:
    file.write(summary)

print(summary)
```

## 4. Practice 2

This example shows how to write, read, and truncate a file using a single open operation.

```python
with open(write_read_path, "w+", encoding="utf-8") as file:
    file.write("Hello world")

    print(file.tell())

    file.seek(0)
    print(file.tell())
    print(file.read())

    file.truncate(5)

    file.seek(0)
    print(file.read())
```

## 5. Exception Handling

Exceptions help us handle errors safely instead of stopping the program unexpectedly. In this section, we use `try`, `except`, `else`, and `finally`.

```python
try:
    numerator = 10
    denominator = 0
    result = numerator / denominator
except ZeroDivisionError:
    print("You cannot divide a number by 0.")

print("hello world")
```

```python
try:
    x = y
except NameError as e:
    print(e)

print("hello world")
```

```python
try:
    with open(sample_path, "r", encoding="utf-8") as file:
        content = file.read()
except Exception as e:
    print(e)
else:
    print(content)
finally:
    print("Cleanup finished.")
```
