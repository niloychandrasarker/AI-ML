## Problem 1

 Write a function introduce(name, age) that prints:

 My name is <name> and I am <age> years old.

Call it using positional arguments.

Sample Input:
Rhim 28

Sample Output:
My name is Rahim and I am 28 years old.





```python
def introduce(name,age):
  print(f"My name is {name} and I am {age} years old.")

name = input("enter your name ")
age = input("enter your age ")

introduce(name,age)
```

## Problem 2

Write a Python generator function called batch_generator. It should take a list of numbers (representing data samples) and a batch_size. The generator should yield one batch at a time as a list.

Example Input:
data = [1, 2, 3, 4, 5, 6, 7, 8]
batch_size = 3

Expected Output (when iterated):
[1, 2, 3]
[4, 5, 6]
[7, 8]







```python
list =  [1, 2, 3, 4, 5, 6, 7, 8]

def data_generator(chunk_size,list):
  for i in range(0, len(list), chunk_size):
    yield list[i:i+chunk_size]

# x = data_generator(3,list)

for batch in data_generator(3,list):
  print(batch)
```

## Problem 3

 Given a list [1,2,3,4], use map() and lambda to create a new list with squares of each number.

Sample Input:
 No input

Sample Output:
[1, 4, 9, 16]






```python
nums = [1, 2, 3, 4]

squares = map(lambda x: x*x, nums)

print(*squares)
print(type(squares))  # unpack kore print
```

## Problem 5 ( Conceptual Session )

 Given a list [1,2,3,4,5,6], use filter() to select even numbers and map() with lambda to square them.

Sample Input:
 No input

Sample Output:
[4, 16, 36]







```python
nums = [1, 2, 3, 4, 5, 6]

even_numbs = filter(lambda x : x%2==0,nums)
squares = map(lambda x:x**2, even_numbs)

print(*squares)
```
